```python
import pandas as pd
stock_prices_dataframe =pd.read_csv('stock_prices.csv')

stock_prices_dataframe
```

```python
twitter_sentiment_dataframe = pd.read_csv('twitter_sentiment.csv')

twitter_sentiment_dataframe
```

```python
twitter_sentiment_dataframe.info()
```

```python
twitter_sentiment_dataframe= twitter_sentiment_dataframe.dropna(axis = 1)

twitter_sentiment_dataframe
```

```python
stock_prices_dataframe= stock_prices_dataframe.dropna(axis = 1)

stock_prices_dataframe
```

```python
merged_dataframe = stock_prices_dataframe.merge(twitter_sentiment_dataframe, on = "Date")
merged_dataframe
```

```python
merged_dataframe.to_csv("merged_stocks_and_twitter_sentiment.csv", index= False)
```
