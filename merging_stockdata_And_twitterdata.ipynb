```python
merged_dataframe = pd.read_csv("merged_stocks_and_twitter_sentiment.csv")

merged_dataframe
```

```python
dataframe = merged_dataframe[["Date","Adj Close", "Volume","ts_polarity","twitter_volume"]]
dataframe
```

```python
dataframe.set_index("Date", inplace = True)

dataframe
```

```python
POSITIVE_SENTIMENT_THRESHOLD= 0.05
NEGATIVE_SENTIMENT_THRESHOLD= -0.05
sentiments = []
for score in dataframe["ts_polarity"]:
  if score>= POSITIVE_SENTIMENT_THRESHOLD:
    sentiments.append("Positive")
  elif score <= NEGATIVE_SENTIMENT_THRESHOLD:
    sentiments.append("Negative")

  else:
    sentiments.append("Neutral")

dataframe["Sentiments"]= sentiments

dataframe
```

```python
dataframe["Sentiments"].value_counts()
```

```python
dataframe["Adj Close"].value_counts()
```

```python
dataframe["Price Difference"] = dataframe["Adj Close"].diff()
dataframe.dropna(inplace = True)
dataframe
```

```python
import numpy
RISE = 1
FALL =0
dataframe["Stock Trend"] = numpy.where(dataframe["Price Difference"]>0, RISE, FALL)
dataframe
```

```python
new_dataframe=dataframe[["Adj Close", "Volume","twitter_volume","Sentiments","Stock Trend"]]
new_dataframe = pd.get_dummies(new_dataframe, columns= ["Sentiments"])
new_dataframe
```

```python

X= new_dataframe.copy()
X.drop("Stock Trend", axis =1, inplace=True) 
X
```

```python
#Random Forest Classifier Algorithm
y = new_dataframe["Stock Trend"].values.reshape(-1,1)
y[:10]
```

```python
SPLIT = int(0.8*len(X))
X_train = X[:SPLIT]
X_test=X[SPLIT:]

y_train = y[:SPLIT]
y_test= y[SPLIT:]
```
